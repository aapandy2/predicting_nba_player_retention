Some attempts at finding missing IDs for the player salary data. 

```python
import pandas as pd
import numpy as np
from unidecode import unidecode
import jellyfish
```

```python
missing_id = pd.read_csv('no_pplayer_id_salaries.csv')
```

```python
missing_id
```

```python
missing_id = missing_id[['Names']]
```

```python
missing_id = missing_id.drop_duplicates()
```

```python
player_names = pd.read_csv('Data/player_identification/player_id_name.csv')
player_names = player_names['NAME'].to_list()
```

Three different functions to attempt matching. Modification of some parameters and/or a smarter function may help. 

```python
def alternate_name_lev(name):
    name = unidecode(name)
    for i in range(0, 4):
        for match_name in player_names:
            if jellyfish.levenshtein_distance(name, match_name) == i:
                return match_name

    return 'no match found'
```

```python
def alternate_name_jaro(name):
    name = unidecode(name)

    for i in range(0, 6):
        for match_name in player_names:
            if jellyfish.jaro_similarity(name, match_name) >= 1-.02*i:
                return match_name

    return 'no match found'
```

```python
def alternate_name_jaro_winkler(name):
    name = unidecode(name)

    for i in range(0, 6):
        for match_name in player_names:
            if jellyfish.jaro_winkler_similarity(name, match_name) >= 1-.02*i:
                return match_name

    return 'no match found'
```

Attempt matching with Levenshtein distance. 

```python
missing_id['alternate_name_lev'] = missing_id['Names'].apply(alternate_name_lev)
```

Attempt matching with Jaro Similarity.

```python
missing_id['alternate_name_jaro'] = missing_id['Names'].apply(alternate_name_jaro)
```

Attempt matching with Jaro-Winkler Similarity.

```python
missing_id['alternate_name_jaro_winkler'] = missing_id['Names'].apply(alternate_name_jaro_winkler)
```
