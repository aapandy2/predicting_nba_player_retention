Testing some models with the data we have. First a few plots with one feature compared between waived and nonwaived (PLAYER-SEASON-TEAMS)'s

```python
#First let's add a columns to the merged data to convert relevant stats to percentages:
merged_df['FG%'] = merged_df['FGM']/merged_df['FGA']
merged_df['3FG%'] = merged_df['FG3M']/merged_df['FG3A']
merged_df['FT%'] = merged_df['FTM']/merged_df['FTA']
cols_to_scale = ['PTS', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF']
merged_df[cols_to_scale] = merged_df[cols_to_scale].div(merged_df['GP'], axis=0)
merged_df.rename(columns={col: f'{col}/G' for col in cols_to_scale}, inplace=True)
```

```python
merged_df.sample()
```

For the moment our rows are (PLAYER-SEASON-TEAM), so we are looking at predictive models that only use a player's performance with their current team to decide if they will be waived. Let's first split the data into a data frame for waived players, and a data frame for nonwaived players. In both we should have some threshold #games to avoid unusually high percentages. 

```python
Games_Threshold = 10
waived_players = merged_df[(merged_df['WAIVED'] == 1) & (merged_df['GP'] > Games_Threshold)]
nonwaived_players = merged_df[(merged_df['WAIVED'] == 0) & (merged_df['GP'] > Games_Threshold)]
```

```python
len(waived_players)
```

```python
len(nonwaived_players)
```

With a 10 game threshold there are only 694 waived players compared to more than 25000 nonwaived players. Is this reasonable to expect? We should keep in mind that the waived player data includes many players who are waived right before a season starts, so there are lots of potential (PLAYER-SEASON-TEAMS) that really are missing because of waivers coming from October waivers. 

For now, let's see a comparison of plots for FG% between both sets of players.

```python

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(waived_players['FG%'].dropna(), bins=20, color='blue', alpha=0.7)
plt.title('FG% for Players waived')
plt.xlabel('Field Goal Percentage')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(nonwaived_players['FG%'].dropna(), bins=20, color='green', alpha=0.7)
plt.title('FG% for Players not waived')
plt.xlabel('Field Goal Percentage')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

This difference between FG% between waived and nonwaived players, is extremely statistically significant:

```python
unstandardizedZ = waived_players.loc[:,'FG%'].mean()
mu = nonwaived_players.loc[:, 'FG%'].mean()
sigma = nonwaived_players.loc[:, 'FG%'].std()
size = len(waived_players)
standardizedZ = (size*unstandardizedZ - mu*size)/(np.sqrt(size)*sigma)
p_value = stats.norm.cdf(standardizedZ)
print(standardizedZ)
print(p_value)
```

Now let's look at other features the same test for other features

```python
unstandardizedZ = waived_players.loc[:,'PTS/G'].mean()
print("average for waived players: ", unstandardizedZ)
mu = nonwaived_players.loc[:, 'PTS/G'].mean()
print("average for nonwaived players: ", mu)
sigma = nonwaived_players.loc[:, 'PTS/G'].std()
size = len(waived_players)
standardizedZ = (size*unstandardizedZ - mu*size)/(np.sqrt(size)*sigma)
p_value = stats.norm.cdf(standardizedZ)
print(standardizedZ)
print(p_value)
```

... and maybe a defensive stat too

```python
unstandardizedZ = waived_players.loc[:,'BLK/G'].mean()
print("average for waived players: ", unstandardizedZ)
mu = nonwaived_players.loc[:, 'BLK/G'].mean()
print("average for nonwaived players: ", mu)
sigma = nonwaived_players.loc[:, 'BLK/G'].std()
size = len(waived_players)
standardizedZ = (size*unstandardizedZ - mu*size)/(np.sqrt(size)*sigma)
p_value = stats.norm.cdf(standardizedZ)
print(standardizedZ)
print(p_value)
```

Now I will rerun the KNN model with this updated data

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
merged_df = merged_df.dropna(subset=['PTS/G', 'REB/G', 'AST/G', 'STL/G', 'BLK/G', 'TOV/G', 'PF/G'])
X = merged_df[['PTS/G', 'REB/G', 'AST/G', 'STL/G', 'BLK/G', 'TOV/G', 'PF/G']]
y = merged_df['WAIVED']
```

```python
len(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=812, stratify = y)
```

```python
knn_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier(n_neighbors = 5))
])
    
```

```python
knn_pipeline.fit(X_train, y_train)
```

```python
y_pred = knn_pipeline.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
accuracy
```

```python
CM = confusion_matrix(y_test, y_pred)
```

```python
CM
```

What about a logistic regression instead?

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(penalty = None)
log_reg.fit(X_train, y_train)
y_pred = log_reg.predict(X_test)
```

```python
accuracy = accuracy_score(y_pred, y_test)
```
